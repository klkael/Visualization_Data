```python
import pandas as pd
import matplotlib.image as im
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('Atlit.csv')
dfcor = df.corr()
plt.imshow(df.corr(),cmap='hot_r')
plt.colorbar()
col = list(map(lambda x:x,df.corr().columns.tolist()))
plt.xticks(np.arange(len(dfcor.index)),dfcor.index.tolist(),rotation=45)
plt.yticks(np.arange(len(dfcor.index)),dfcor.index.tolist())
for x in range(len(col)):
    for y in range(len(col)):
        plt.text(y,x,round(dfcor.iloc[x,y],4),color='blue',ha='center',va='center') #angka 4 buat set berapa angka belakang koma
plt.xlim(-.5,4.5)
plt.ylim(-.5,4.5)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d.axes3d

fig = plt.figure()
p = plt.subplot(111,projection='3d')
data = range(11)
x = np.array(data)
x

p.plot_wireframe(x,x,x.reshape(1,-1),color='r',linestyle=':')
p.scatter(x,x,x,marker='*',color='y',s=150)
plt.show()
```

```python
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
plt.figure(figsize=(10,10))
p = plt.subplot(111,projection='3d')
x1 = np.array([6,0,0])
y1 = np.array([0,-3,0])
z1 = np.array([0,0,6])
p.scatter(x1,y1,z1,marker='*',s=110)
verts = [list(zip(x1,y1,z1))]
srf = Poly3DCollection(verts,alpha=.25,facecolors='red')
plt.gca().add_collection3d(srf)
plt.show()
```

```python
plt.figure(figsize=(10,10))
p = plt.subplot(111,projection='3d')
x2 = np.array([1.33,0,0])
y2 = np.array([0,4,0])
z2 = np.array([0,0,-2])
p.scatter(x2,y2,z2,marker='$halo$',s=710)
verts = [list(zip(x2,y2,z2))]
srf = Poly3DCollection(verts,alpha=.25,facecolors='green')
plt.gca().add_collection3d(srf)
plt.show()
```
