```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_excel('indo_12_1.xlsx',na_values=['-'],header=3,skipfooter=3)
df.rename(columns={'Unnamed: 0':'Provinsi'},inplace=True)
df = df.fillna(0)
tahun = []
for i in df.columns[1::]:
    tahun.append(i)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.mplot3d.axes3d
fig = plt.figure('3d',figsize=(16,16))
p = plt.subplot(111,projection='3d')
plt.suptitle('Penduduk Indonesia', size=25)
for i in range (len(tahun)):
    x = np.arange(len(df['Provinsi']))
    y = [i] * len(df['Provinsi'])
    z = np.zeros(len(df['Provinsi']))
    dx = np.ones(len(df['Provinsi']))
    dy = np.ones(len(df['Provinsi']))
    dz = df[tahun[i]]
    p.bar3d(x,y,z,dx,dy,dz)
p.set_xticks(np.arange(len(df['Provinsi'])))
p.set_xticklabels(df['Provinsi'],rotation=90)
p.set_yticks(np.arange(len(tahun)))
p.set_yticklabels(tahun)
plt.show()
```

```python
fig = plt.figure('3d',figsize=(16,16))
p = plt.subplot(111,projection='3d')
plt.suptitle('Penduduk Indonesia Tahun 1971',size=25)

x = np.arange(len(df['Provinsi']))
y = np.arange(len(df['Provinsi']))
z = np.zeros(len(df['Provinsi']))
dx = np.ones(len(df['Provinsi']))
dy = np.ones(len(df['Provinsi']))
dz = df[1971]
p.bar3d(x,y,z,dx,dy,dz,color='maroon')
p.set_xticks(np.arange(len(df['Provinsi'])))
p.set_xticklabels(df['Provinsi'],rotation=90)
plt.show()
```
