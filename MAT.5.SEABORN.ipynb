```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn

plt.style.use('seaborn')
x = np.arange(10)
y = np.arange(10)
plt.plot(x,y)
plt.show()
#plt.style.available
```

```python
df = pd.DataFrame({
    'x':np.arange(10),
    'y':np.arange(10),
    'z':['Pria','Wanita']*5
})
plt.style.use('seaborn')
plt.plot(df.x,df.y)
```

```python
import seaborn as sb
sb.set(style='whitegrid')
sb.relplot(data=df, x='x', y='y',hue='x',style='z',kind='line',size='z') #kind='line'
sb.barplot(data=df, x='x', y='y')
```
