```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'x':[1,2,3,4,5],
    'y':[9,5,7,3,6],
    'z':[2,3,4,1,3]
})
plt.figure('Tes',figsize=(16,8))
plt.plot(df['x'],df[['y','z']])
plt.grid()
plt.title('Tes')
plt.xticks(np.arange(6),list('abcdef'))
plt.yticks(np.arange(0,11),rotation=90)
plt.show()
```

```python
df = pd.DataFrame({
    'x':[1,2,3,4,5],
    'y':[9,5,7,3,6],
    'z':[2,3,4,1,3]})

plt.figure('Tes')
plt.suptitle('super title')
plt.subplot(221)
plt.plot([1,2,3],[1,2,3])
plt.grid()
plt.title('1')

plt.subplot(224)
plt.plot([1,2,3],[1,2,3])
plt.grid()
plt.title('4')
```

```python
plt.bar(df.x,df.y,color='pink',edgecolor='black',yerr=df.describe().loc['std']['y'])
plt.bar(df.x,df.z,color='black')
plt.title('Bar chart',fontdict={'family':'impact','size':30,'color':'green'})
plt.xlabel('ini x',fontsize=20)
plt.ylabel('ini y',fontsize=20)
plt.yticks(np.arange(0,11))

for i in df['y']:
    plt.text(df[df['y']==i]['x'],i,i)
plt.show()
```

```python
plt.barh(df.x,df.y,color='pink',edgecolor='black')
plt.barh(df.x,df.z,color='black')
plt.title('Bar chart',fontdict={'family':'impact','size':30,'color':'green'})
plt.xlabel('ini x',fontsize=20)
plt.ylabel('ini y',fontsize=20)

for i in df['y']:
    plt.text(i,df[df['y']==i]['x'],i)
plt.show()
```

```python
plt.scatter([1,2,3],[4,5,6],marker='X',s=300,color='y',zorder=2)
plt.plot([1,2,3],[4,5,6],markersize=30,zorder=1)
plt.show()
```
